#### 1 - Расчет длинв вектора в пространстве

```python
%matplotlib inline
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def count_vect(a,b):
    if len(a) == 3 and len(b) == 3:
        x1,x2,x3 = a[0],a[1],a[2]
        y1,y2,y3 = b[0],b[1],b[2]
        result = math.sqrt((y1-x1)**2 + (y2-x2)**2 + (y3-x3)**2)   
        return result
    else:
        print('No solution')
```

```python
a = [10,10,10]
b = [0,0,-10]
```

```python
count_vect(a, b)
```

#### 3 Полярные и декартовые координаты

```python
# Перевод полярных координат в декартовы
def pol2cart(r, f):
    x = r * np.cos(f)
    y = r * np.sin(f)
    return x, y
```

```python
# График окружности
x = []
y = []
y1 = []
c = 5
for i in range(0, 360):
    h = [pol2cart(c, i)]
    x.append(h[0][0])
    y.append(h[0][1])
    y1.append(-h[0][1])
plt.plot(x,y)
plt.plot(x,y1)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```python
# График прямой
def polar_line(R=1, a=45):
    f = np.array([np.radians(a), np.radians(a)])
    r = np.array([0, R])
    plt.polar(f, r)
    plt.show()
polar_line(R=6, a=30)
```

#### 4 - решить уравнение

```python
x = np.linspace(-2, 3, 201)
plt.plot(x, (np.exp(x) -1+ x)/x)
plt.plot(x, x**2-1)
```

```python
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y-x**2 + 1, np.exp(x) +x*(1-y) - 1)
x1, y1 = fsolve(equations, (2,4))

print(x1, y1)
```
